```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#0 ile 1000 arasında 500 random sayı olustur
x_kordinat=np.random.randint(0,1000,500)
print(x_kordinat)
```

```python
##0 ile 1000 arasında 500 random sayı olustur
y_kordinat=np.random.randint(0,1000,500)
print(y_kordinat)
```

```python
#Olusturulan sayilari data frame aktar
data = {
    'X kordinatları': x_kordinat,
    'Y kordinatları': y_kordinat
}
dataFrame= pd.DataFrame(data)
dataFrame
```

```python
#DataFrame excel dosyasına kaydet
dataFrame.to_excel('KordinatTablosu.xlsx', index=False)
```

```python
#excel doysasındaki sutunları oku
excel_yol = 'KordinatTablosu.xlsx'
dRead = pd.read_excel(excel_yol)
x_kordinatRead = dRead['X kordinatları']
y_kordinatRead = dRead['Y kordinatları']
print(x_kordinatRead)
print(y_kordinatRead)
```

```python
#DataFrame olustur
dataTable= {
    'x_kordinat': x_kordinatRead,
    'y_kordinat': y_kordinatRead
}
df = pd.DataFrame(dataTable)

#Grid boyutu ve grid sayısını belirle
grid_size = 200
num_grids = 1000 // grid_size

#Renk haritası oluştur
colors = plt.cm.rainbow(np.linspace(0, 1, num_grids**2))

#Grafiğin boyutunu ayarla
plt.figure(figsize=(10, 10))


for i in range(num_grids):
    for j in range(num_grids):
        #grid x ve y eksenlerinde başlangıç ve bitiş indekslerini bul
        x_min = i * grid_size
        x_max = (i + 1) * grid_size
        y_min = j * grid_size
        y_max = (j + 1) * grid_size

        #belirtilen grid noktalarını şeç
        mask = (df['x_kordinat'] >= x_min) & (df['x_kordinat'] < x_max) & \
               (df['y_kordinat'] >= y_min) & (df['y_kordinat'] < y_max)

        #Plot grafğini çiz
        plt.scatter(df[mask]['x_kordinat'], df[mask]['y_kordinat'], color=colors[i * num_grids + j], label=f'Grid {i},{j}')

plt.xlabel('X kordinatları')
plt.ylabel('Y kordinatları')
plt.grid(True)

plt.show()
```
